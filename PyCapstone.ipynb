The purpose of this data analysis is to find the differences between citibike causal riders and members in 2023. Though users can purchase single rides, day passes and annual memberships, as of 2023 single riders and day pass user have been combined and labled as **casual** users in the data. Paying attention to the differences between casual and members can hopefully give insight into how to go about marketing to casual users, as it may be more profitable for users to be annual members than casual riders.

```python
#since each dataset has the same column names and datatypes for each corresponding column
#will create a dataframe to combine the datasets for 2023

df = pd.concat([df01, df02, df03, df04, df05, df6, df7, df8, df9, df10, df11, df12], ignore_index = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
#creating a column for trip duration 

df['duration'] = pd.to_datetime(df['ended_at']) - pd.to_datetime(df['started_at'])
```

```python
df.head()
```

```python
# getting trip duartion in minutes 
#according to citibike any trips that were below 60 seconds in length has been removed
#because those trips were potentially false starts or users trying to re-dock a bike to ensure it's secure

df['trip_duration']= df['duration'].dt.total_seconds()/60.00
```

```python
df.head()
```

```python
df1 = df.round({'trip_duration': 0})
```

```python
df1.head()
```

```python
#dropping the duration column as it no longer necessary

df1.drop('duration', axis=1, inplace=True)
```

```python
df1.head()
```

```python
# planning to extract month into it's own column so need start time to be a datetime object

df1['started_at']= pd.to_datetime(df['started_at'])

#df1['ended_at'] = pd.to_datetime(df['ended_at'])  
```

```python
df1.dtypes
```

```python
#extracting month into it's own column

df1['month'] = pd.DatetimeIndex(df['started_at']).month
```

```python
df1.tail()
```

```python
#extract day of the week when trip was taken and create a column for it

df1['day_of_week'] = df1['started_at'].dt.day_name()
```

```python
#going to check the max in minutes trip duration (check for oddities, and make sure no trip are a day or longer)
#before dropping the start and end time columns 
#since start day and end day of a trip should be the same day 
#(except for those that start at midnight and end the next day)

df1['trip_duration'].max()

```

```python
#there seem to be outliers in the data, as the median trip is 6 min long
#and according to citibike trips are included for the first 30 min for casual user
#and 45 min for members, every min after is charged for both

td_bplt = df1.boxplot(column = 'trip_duration')


td_bplt.plot()

plt.title('Citibike Trip Durations in Min for 2023')

plt.show()
```

```python
#since a citibike is considered stolen after 24 hours, we will remove all trips goingover 24 hrs
#creating a new dataframe 

df2 = df1[df1['trip_duration'] <= 1440]
```

```python
#making sure trip durations does not exceed 24 hr = 1440 min

df2['trip_duration'].max()
```

```python
#dropping started_at, ended_at and ride id columns
#ride id is completely arbitary and thus not relevant 

df2.drop(columns=['ride_id','started_at', 'ended_at'], axis=1, inplace=True)
```

```python
# How are length of trips affected by month for casual and annual riders

#calculate mean and standard deviation of trip durations for each month
df3_grouped = (
   df3[['trip_duration', 'month_name']].groupby(['month_name'])
   .agg(['mean', 'std', 'count'])
)
df3_grouped = df3_grouped.droplevel(axis=1, level=0).reset_index()

# Calculate a confidence interval as well.
df3_grouped['ci'] = 1.96 * df3_grouped['std'] / np.sqrt(df3_grouped['count'])
df3_grouped['ci_lower'] = df3_grouped['mean'] - df3_grouped['ci']
df3_grouped['ci_upper'] = df3_grouped['mean'] + df3_grouped['ci']
df3_grouped.head()
```

```python
#By default, if there are more than one observations for each unique x value,
#Seaborn will take the mean of those observations and bootstrap a 95% confidence interval 
#using 1000 iterations

#set graph size 
plt.figure(figsize=(12,8))

# Set the style
sns.set_style("darkgrid")

r5 = sns.lineplot(data=df3, x='month_name', y='trip_duration', hue ="member_casual", palette = "CMRmap")
# A more verbose but explicit version of the above not relying on defaults:
# ax = sns.lineplot(data=df, x='pickup_date', y='fare', estimator='mean', ci=95,
#                   n_boot=1000)
r5.set_title('Trip Duration by Month in 2023')
r5.set_ylim(0)
plt.xlabel('Month', labelpad = 20, fontsize=15)
plt.ylabel('Trip Duration',labelpad = 20, fontsize=15)
plt.xticks(fontsize = 11) 
plt.yticks(fontsize = 12) 
legend_handles, _= r5.get_legend_handles_labels()
r5.legend(legend_handles, ['Annual Member','Casual Rider'], 
         bbox_to_anchor=(1,1), 
       title='Type of Rider', fontsize="15", title_fontsize=15)
#r5.figure.autofmt_xdate(rotation=45)

plt.show()
#plt.savefig('my_plot6.png')
```
