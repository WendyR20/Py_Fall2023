## Conclusion

Using Grover Cleveland High School as an initial comparison, the aim was to compare U.S. History Regents scores to that of other High Schools in the borough of Queens, more specifially we will be looking at how the mean score and percentage of students scoring higher than 65 compares. 

For the 2015, 2016 and 2017 year, both the mean score and percentage of studnets scoring higher than 65 was lower for Grover Cleveland than for students across Queens High Schools on the whole.

The Mean Score and percentage scoring higher than 65 for Grover Cleveland students was lower than that of students across Queens High Schools by a range of 3-5 points (mean score), and a range of 5-8.5%(scoring higher than 65) for the years 2015, 2016, an 2017.
