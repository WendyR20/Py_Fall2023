```python
#Comparing the data for time spent in hospital for Caucasians to that of other categories
#need to first drop all rows with '?', cannot determine what was meant to be there

Rdf = df[df['race'] != '?']
Rdf.groupby(['race'])['race'].count().sort_values(ascending = False)
```

```python
#subsetting data, creating df for only race and time-in-hospital
#including number of diagnosis out of curiosity

RTdf = Rdf[['race','time_in_hospital','number_diagnoses']]
```

```python
#also renaming the columns as time-in-hospital could be more accurate by saying days

RT2df = RTdf.rename(columns={'race':'Race', 'time_in_hospital':'Days_In_Hospital', 'number_diagnoses':'Num_of_Diagnoses'})
```

```python
RT2df.head()
```

```python
#adding a space so AfricanAmerican reads African American

RT2df['Race'].replace('AfricanAmerican','African American',inplace=True)
```

```python
#boxplot comparing data across races for days in hospital

palette = sns.color_palette("husl", 8)
sns.boxplot(x='Race', y='Days_In_Hospital', data= RT2df, palette = palette)
```

```python
#pairplots to see how the data compares 

sns.pairplot(RT2df, hue = 'Race')
```

```python
#want to take a look at the pair plots w/o the Caucasian data, 
#as there is so much of it compared to that of other races, it obscures (overwhelms?) the rest

RT3df = RT2df[RT2df['Race'] != 'Caucasian']
RT3df.head()
```

```python
# pair plots again

sns.pairplot(RT3df, hue = 'Race')
```

```python
#now looking only at data for the classfications: other, asian and hispanic 
#again there is much more data for African Americans, 
#and while it does not visually obfuscate the rest of the data as much as Caucasian data did
#it would be interesting to see visualizations for the other data

RT4df = RT3df[RT2df['Race'] != 'African American']
RT4df.head()
```

```python
#pair plots, third time's the charm

sns.pairplot(RT4df, hue = 'Race')
```

```python
#seeing if there's any correalation between number of dianoses and time spent in hospital

sns.lmplot(x = "Num_of_Diagnoses", y = "Days_In_Hospital",
           hue = "Race", data = RT2df) 
```

```python
#seperate graph for each race, since Num_of_Diagnoses is discrete
#trying to see if it's a weak positve correalation or not

sns.lmplot(x = "Num_of_Diagnoses", y = "Days_In_Hospital",
           col = "Race",
           hue = "Race", 
           palette = "Set2",
           data = RT2df)
```

## Conclusion

Using the **Caucasian** group data as an initial comparison, the aim was to compare the amount of days spent in hospital to that of thr **African-American**, **Hispanic**, **Asian** and **Other** groups. 

While the average (mean) days spent in hospital for the Caucasian group was 	**4.385721 days**, this was the second highest average, only second to the African-American group. The median days spent in hospital for the Caucasian group was **4 days**, the same as the African-American group but higher than the median of the Asian, Other and Hispanic groups. When using a box plot to compare the data across groups, it is visually evident that the Hispanic and Asian group data appears to have more outliers than the data for the Caucasian, African-American and Other groups.
